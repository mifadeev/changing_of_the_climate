### Подготовка и очистка данных

**Проверяем данные на нулевые значения:**

```python
data.isnull().sum()
```

Заменяем отсутствующие значения на средние:

```python
data.iloc[:, 10:72].head()
```

```python
data_columns_list = list(data.columns)
columns_fill = data_columns_list[10:]
for year in columns_fill:
    data[year] = data[year].fillna(data[year].mean())
```

```python
data.iloc[:, 10:].isnull().sum()
```

Изменим название столбцов  с годами наблюдения данных. Уберем буквенный литерал, оставим только число(год)

```python
data.columns = data.columns.str.replace('F', '')
```

```python
data.columns
```

**Проверяем данные на наличие дубликатов**

```python
data.duplicated()
```

Дублирующие данные в таблице отсутствуют.

## Анализ данных

Проанализируем изменение среднее значение отклонения от базовой температуры по годам.

```python
years_list = list(data.columns[10:])
years_list_int = [int(i) for i in years_list]
mean_temperature = data[years_list].mean()

plt.figure(figsize = (20, 10))
plt.plot(years_list_int, mean_temperature, marker = 'o', color = 'green')
plt.title('Изменение глобальной средней температуры по годам', fontsize = 18)
plt.xlabel('Год', fontsize = 14)
plt.ylabel('Среднее значение отклонения температуры от базовой, (°C)', fontsize = 14)
plt.grid()
plt.show()
```

Среднее изменение температуры от базового значения с годами увеличивается.

Рассмотрим максимальные значения по годам наблюдений

```python
# берём данные и предыдущего расчета
# years_list = list(data.columns[10:])
# years_list_int = [int(i) for i in years_list]

max_temperature = data[years_list].max()

plt.figure(figsize = (20, 10))
plt.plot(years_list_int, max_temperature, marker = 'o', color = 'red')
plt.title('Изменение глобальной максимальной температуры по годам', fontsize = 18)
plt.xlabel('Год', fontsize = 14)
plt.ylabel('Максимальное значение отклонения температуры от базовой, (°C)', fontsize = 14)
plt.grid()
plt.show()
```

Минимальное значение по годам наблюдений

```python
# берём данные и предыдущего расчета
# years_list = list(data.columns[10:])
# years_list_int = [int(i) for i in years_list]

min_temperature = data[years_list].min()

plt.figure(figsize = (20, 10))
plt.plot(years_list_int, min_temperature, marker = 'o', color = 'blue')
plt.title('Изменение глобальной минимальной температуры по годам', fontsize = 18)
plt.xlabel('Год', fontsize = 14)
plt.ylabel('Минимальное значение отклонения температуры от базовой, (°C)', fontsize = 14)
plt.grid()
plt.show()
```

Для наглядности рассмотрим полученные данные о максимальных и минимальных значениях вместе

```python
plt.figure(figsize = (20, 10))
plt.bar(years_list_int, min_temperature, color = 'blue', label = 'Минимальное отклонение от базовой температуры')
plt.bar(years_list_int, max_temperature, color = 'red', label = 'Максимальное отклонение от базовой температуры')
plt.title('Изменение глобальной минимальной и максимальной температуры по годам', fontsize = 18)
plt.xlabel('Год', fontsize = 14)
plt.ylabel('Значение отклонения температуры от базовой, (°C)', fontsize = 14)
plt.grid()
plt.legend()
plt.show()
```

## Выводы

Происходит плавное постепенное изменение температуры в сторону увеличения значения от принятой базовой температуры.
Наибольший рост стал заметен начиная с 1990 года и по настоящее время.
Самые незначительные изменения наблюдались с 1960 по 1980 и колебались около нулевого отклонения то в большую то в меньшую сторону. С 1990 года начинается значительный рост именно положительного отклонения температуры от базового значения.
